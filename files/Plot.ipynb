```python
#y=x^2  from -3 to 3 and fit line with noise of SD .5 and plot it
x1=np.random.random((500))*(-3)
x1
```

```python
x2=np.random.random((500))*(3)
x2
```

```python
arr=np.append(x1,x2)
```

```python
arr.shape
```

```python
mactual1, cactual1=2,3
noise1=np.random.randn(1000)
y1=arr*arr+noise1*0.5
plt.scatter(arr,y1)
#plt.scatter(arr,y1) # do not use when SD is more
```

```python
reg=LinearRegression()
arr=arr.reshape(-1,1)
reg.fit(arr,y1)
```

```python
bias=reg.intercept_
bias
```

```python
cof=reg.coef_[0]
cof
```

```python
dummyx=np.linspace(-3,3,2000)
dummyy=cof*dummyx+bias
plt.scatter(arr,y1)
plt.plot(dummyx,dummyy)
```
