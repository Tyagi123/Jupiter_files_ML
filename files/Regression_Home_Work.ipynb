```python
import numpy as np
import pandas as pd 
```

```python
x1=np.random.random((500))*(-3)
```

```python
x2=np.random.random((500))*(3)
```

```python
x1=np.append(x1,x2)
```

```python
x3=np.random.random((500))*(-3)
x4=np.random.random((500))*(3)
x2=np.append(x3,x4)
```

```python
noisearg=np.random.randn()*0.5
```

```python
noisearg
```

```python
y=2*x1+3*x2+noisearg
```

```python
from sklearn.linear_model import Ridge 
```

```python
rd= Ridge()
```

```python
y.shape
```

```python
x1.shape
```

```python
dataset = pd.DataFrame({'x1': x1, 'x2': x2}, columns=['x1', 'x2'])
```

```python
rd.fit(dataset,y)
```

```python
print(rd.alpha)
print(rd.coef_)
print(rd.intercept_)
```

```python
from sklearn.linear_model import Lasso
ls=Lasso()
```

```python
ls.fit(dataset,y)
print(ls.alpha)
print(ls.coef_)
print(ls.intercept_)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(dataset,y)
print(lr.coef_)
print(lr.intercept_)
```
